```python
import pandas as pd
import matplotlib.pyplot as plt
df = pd.read_csv("/content/StudentsPerformance.csv")
avg_df = pd.DataFrame({
    "Subject": ["Math", "Reading", "Writing"],
    "Average_Marks": [
        df["math score"].mean(),
        df["reading score"].mean(),
        df["writing score"].mean()
    ]
})
plt.figure(figsize=(8, 6))
plt.bar(avg_df['Subject'], avg_df['Average_Marks'], color=['skyblue', 'lightgreen', 'salmon'])
plt.xlabel("Subjects")
plt.ylabel("Average Marks")
plt.title("Average Marks of Students in Different Subjects")
plt.show()
print("Easiest Subject:", avg_df.loc[avg_df["Average_Marks"].idxmax(), "Subject"])
print("Toughest Subject:", avg_df.loc[avg_df["Average_Marks"].idxmin(), "Subject"])
```

```python
import pandas as pd
import matplotlib.pyplot as plt
df = pd.read_csv("/content/Sample - Superstore.csv", encoding="latin1")
plt.figure(figsize=(8,6))
plt.bar(df['Category'], df['Sales'], color=plt.cm.Paired.colors)
plt.xlabel("Product Category")
plt.ylabel("Sales")
plt.title("Sales Distribution by Product Category")
plt.xticks(rotation=45)
plt.show()
best_category = df.loc[df["Sales"].idxmax(), "Category"]
print("Best-Selling Category:", best_category)
```
